```python
#importing Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Reading dataset 

telco_data=pd.read_excel('Telco-Customer-Churn.xlsx')
telco_data.head()
```

# 1. Compare churn count with respect to gender

```python
telco_data.columns              #To show the column names
```

```python
churn_count=telco_data['Churn'].value_counts()              #to find the total churn count
churn_count
```

```python
sns.countplot(x='Churn',hue='gender',data=telco_data)      # To display the churn count with respect to gender
```

# 2. Find out how many female senior citizens there in the dataset

```python
fem_dat=telco_data[telco_data['gender']=='Female']    # To fetch all the records of gender female 
```

```python
female_count=fem_dat.value_counts().sum()  # To show all Female Counts
female_count
```

```python
female_data=telco_data[(telco_data['gender']=='Female') & (telco_data['SeniorCitizen']==1.0)]  # To fetch all records of female
female_data                                                                                  # senior citizens
```

```python
female_data_count=female_data.value_counts().sum()                 # Female senior citizen count
female_data_count
```

```python
# To display the count plot of Female Senior citizens

sns.countplot(x=fem_dat['gender'],hue='SeniorCitizen',data=fem_dat,palette=['#6E8B3D',"#FFB90F"])
plt.show()
```

# 3. Compare 'tenure' with 'Total Charges'

```python
# Scatter plot to compare tenure with Total Charges

plt.figure(figsize=(8,6))
plt.scatter(telco_data['tenure'],telco_data['TotalCharges'],s=12,marker='^')
plt.title('Plot of tenure Vs Total Charges',size=16)
plt.xlabel('tenure')
plt.ylabel('Total Charges')
plt.tight_layout()
plt.grid()
```

```python
#line plot to compare tenure with Total Charges
sns.lineplot(data=telco_data, x="tenure", y="TotalCharges",color='maroon').set(title='Line Plot of Tenure vs TotalCharges')
```

# 4. Find out which contract is preferred by the senior citizen

```python
#To fetch all the records of Senior Citizens

senior_data=telco_data[telco_data['SeniorCitizen']==1.0]
senior_data.head()
```

```python
# grouping the senior citizen data based on contract

sendat=senior_data.groupby(by=['Contract']).sum()
sendat['SeniorCitizen']
```

```python
# Pie chart to display which contract is preferred by SeniorCitizen

sendat.plot.pie(y='SeniorCitizen', figsize=(7, 7))
# show plot
plt.legend(title='contract preferred by senior citizens')
plt.show()
```

# 5. Comment on your finds on Payment Method

# 5.1 Pie plot of Payment Method which is popular among SeniorCitizens

```python
# Data grouped by PaymentMethod

payment_method=telco_data.groupby(by=['PaymentMethod']).sum()
payment_method
```

```python
#pie plot to display payment method popular among senior citizens

payment_method.plot.pie(y='SeniorCitizen', figsize=(8, 8))
# show plot
plt.legend(title='Payment Method popular among SeniorCitizens')
plt.show()
```

# 5.2  Pie plot to display Total charges in Different Payment Methods

```python
# To display the Pie plot of Total charges in different payment methods

payment_method.plot.pie(y='TotalCharges', figsize=(8, 8))
# show plot
plt.legend(title='TotalCharges in Different Payment Methods')
plt.show()
```

```python
# importing libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Reading the Dataset

country_data=pd.read_excel('Dataset_w6.xlsx')
country_data.head()
```

# 1. Find out how many countries are there in the dataset.

```python
# To find the total number of countries in the dataset

#to find the unique count
country_count=country_data['country'].nunique()

#To display the total number of countries
print('Total Number of Countries in the dataset is :',country_count)
```

# 2. Examine the data by using visualizations.

# 1.Region vs GDP Per Capita

```python
# Grouping the data based on Region
region_gdp=country_data.groupby(by=['Region']).sum()

#Filter only the GDP per capita column value for each Region
filter_gdp=region_gdp['GDP per capita (current US$)']

filter_gdp
```

```python
# Bar graph to display Region vs GDP Per Capita

filter_gdp.plot(kind="bar",figsize=(8, 6))
plt.title("Region vs GDP per capita ")
plt.xlabel("Region")
plt.ylabel("GDP per capita")
```

# Comment :GDP Per Capita is Higher for NorthernEurope Region and Lower for WesternAfrica Region

# 2.Region vs Population

```python
# Grouping the data based on Region
region_pop=country_data.groupby(by=['Region']).sum()

#Filter only the GDP per capita column value for each Region
filter_pop=region_pop['Population in thousands (2017)']

filter_pop
```

```python
# Bar graph to display Region vs Population in Thousands
filter_pop.plot(kind="bar",figsize=(8, 6))
plt.title("Region vs GDP per Population in Thousands")
plt.xlabel("Region")
plt.ylabel("Population in Thousands")
```

```python
# Comment:Population is more in SouthernAsia Region
```

## 3. Find out how many countries with respect to the region are available
## in the dataset

```python
#Finding the Regionwise count of records to find the number of countries
region_data=country_data.groupby(by=['Region']).size()
region_data
```

```python
#To Plot a graph which shows the number of countries for each region

region_data.plot(kind="bar",figsize=(10, 8))
plt.title("Number of countries for each Region")
plt.xlabel("Region")
plt.ylabel("Number of Coutries")
```

```python
#Comment :The Number of countries are high in WesternAsia and lower in Oceania
```

# 4. On average which region has the highest a. GDP per capita (current US$)


```python
#Grouping the data based on Region
region_mean_gdp=country_data.groupby(by=['Region']).mean().round(2)
filter_mean_gdp=region_mean_gdp['GDP per capita (current US$)']

#Sorting to findout the region with highest Average GDP Per Capita
filter_mean_gdp=filter_mean_gdp.sort_values(ascending=False)
filter_mean_gdp
```

```python
#Plot to show the region vs Average GDP Per Capita

filter_mean_gdp.plot(kind="bar",figsize=(10, 8))
plt.title("Average GDP Per capita for each Region")
plt.xlabel("Region")
plt.ylabel("Average GDP Per capita")
```

```python
#Average GDP Per Capita is higher for NorthAmerica Region
```

# On average which region has the highest b. International trade: Imports (million US$)


```python
#Grouping the data based on Region

region_mean_trade=country_data.groupby(by=['Region']).mean().round(2)

filter_mean_trade=region_mean_trade['International trade: Imports (million US$)']

#Sorting to findout the region with highest Imports
filter_mean_trade=filter_mean_trade.sort_values(ascending=False)
filter_mean_trade
```

```python
#Plot to show the region vs Average InternationalTrade

filter_mean_trade.plot(kind="bar",figsize=(10, 8))
plt.title("Average International Trade for each Region")
plt.xlabel("Region")
plt.ylabel("Average International Trade-Imports")
```

```python
#Average International Trade is higher for NorthAmerica Region
```

# 5. Find out which region has more consistent with respect to GDP per capita (current US$)

```python
#Grouping the data based on Region
region_stddev_gdp=country_data.groupby(by=['Region']).std().round(2)
filter_stddev_gdp=region_stddev_gdp['GDP per capita (current US$)']

#Standard Deviation values for each Region
filter_stddev_gdp=filter_stddev_gdp.sort_values()
filter_stddev_gdp
```

```python
filter_stddev_gdp.plot(kind="bar",figsize=(10, 8))
plt.title("Standard Deviation for GDP Per capita for each Region")
plt.xlabel("Region")
plt.ylabel("Standard Deviation for GDP Per capita")
```

```python
# Standard Deviation for GDP per capita  is low for WesternAfrica Region,it shows that WestrnAfrica region is more consistent with respect to 
 #GDP per capita
```
